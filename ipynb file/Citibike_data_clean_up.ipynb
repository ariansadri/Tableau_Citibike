```python
jan = pd.read_csv("CSVFILES/2018-january-citibike.csv")
feb = pd.read_csv("CSVFILES/2018-february-citibike.csv")
mar = pd.read_csv("CSVFILES/2018-march-citibike.csv")
apr = pd.read_csv("CSVFILES/2018-april-citibike.csv")
may = pd.read_csv("CSVFILES/2018-may-citibike.csv")
jun = pd.read_csv("CSVFILES/2018-june-citibike.csv")
jul = pd.read_csv("CSVFILES/2018-july-citibike.csv")
aug = pd.read_csv("CSVFILES/2018-august-citibike.csv")
sep = pd.read_csv("CSVFILES/2018-september-citibike.csv")
october = pd.read_csv("CSVFILES/2018-october-citibike.csv")
nov = pd.read_csv("CSVFILES/2018-november-citibike.csv")
dec = pd.read_csv("CSVFILES/2018-december-citibike.csv")
```
